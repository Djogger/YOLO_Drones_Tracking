```python
from ultralytics import YOLO
import cv2

# Загрузка модели YOLO
model = YOLO('runs/detect/train/weights/best.pt')  # Файл весов 'best.pt'

# Путь к входному видео
video_path = 'test8.mp4'
output_video_path = 'output_video.mp4'

# Открытие видео
cap = cv2.VideoCapture(video_path)

# Получение параметров видео
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# Создание объекта для записи размеченного видео
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Обработка видео по кадрам
frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Выполнение предсказания YOLO на текущем кадре
    results = model.predict(source=frame, conf=0.65, iou=0.3)

    # Получение размеченного кадра
    annotated_frame = results[0].plot()  # Генерация кадра с bounding boxes

    # Сохранение размеченного кадра в выходное видео
    out.write(annotated_frame)

    # Отображение текущего кадра (опционально)
    cv2.imshow('YOLO Video Annotation', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Нажмите 'q' для выхода
        break

    frame_count += 1
    print(f"Processed frame {frame_count}")

# Освобождение ресурсов
cap.release()
out.release()
cv2.destroyAllWindows()

print("Видео успешно размечено и сохранено в:", output_video_path)
```

```python
from ultralytics import YOLO
import cv2

# Загрузка модели
model = YOLO('runs/detect/train/weights/best.pt')

# Выполнение предсказания
results = model.predict(source='test_image1.jpg', conf=0.8, iou=0.25)

# Обработка первого результата
result_image = results[0].plot()  # Генерация изображения с разметкой

# Уменьшение размера изображения (например, до 50% от оригинального размера)
scale_percent = 50  # Процент от оригинального размера
width = int(result_image.shape[1] * scale_percent / 100)
height = int(result_image.shape[0] * scale_percent / 100)
dim = (width, height)

# Изменение размера изображения
resized_image = cv2.resize(result_image, dim, interpolation=cv2.INTER_AREA)

# Отображение уменьшенного изображения
cv2.imshow("Detections", resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
